```python
import pandas as pd
import numpy as np
```

```python
df_user = pd.read_excel('user penetration.xlsx')
df_user.head()
```

```python
df_market = pd.read_excel('market size.xlsx')
df_market.head()
```

```python
df_pop = pd.read_csv('Country_population.csv')
df_pop.drop(['Country Code'], axis=1, inplace=True)
df_pop.head()
```

```python
df_edu = pd.read_excel('education index.xlsx')
df_edu['Country'] = df_edu['Country'].str.strip()
df_edu.head()
```

```python
df_freedom = pd.read_excel('freedom index.xlsx')
df_freedom.drop(['Unnamed: 0'], axis=1, inplace=True)
df_freedom.head()
```

```python
df_immi = pd.read_csv('immigration-by-country-2024.csv')
df_immi.head()
```

```python
df_int = pd.read_excel('internet_penetration_rate.xlsx')
df_int['Country'] = df_int['Country'].str.strip()
df_int.drop(['Region','Internet users','Population'], axis=1, inplace=True)
df_int.head()
```

```python
df_netflix = pd.read_csv('Netflix_Library_Cost_GDP.csv')
df_netflix.head()
```

```python
df_gdp = df_netflix[['Country', 'Total.Library.Size', 'g2020', 'g2021', 'g2022', 'g2023', 'Cost']]
df_gdp.head()
```

```python
# Merge
df_merged = pd.merge(df_gdp, df_edu, on='Country', how='left')
df_merged.head()
```

```python
df_merged = pd.merge(df_merged, df_freedom, on='Country', how='left')
df_merged.head()
```

```python
df_merged = pd.merge(df_merged, df_pop, on='Country', how='left')
df_merged.head()
```

```python
df_merged = pd.merge(df_merged, df_immi, on='Country', how='left')
df_merged.head()
```

```python
df_merged = pd.merge(df_merged, df_int, on='Country', how='left')
df_merged.head()
```

```python
df_merged = pd.merge(df_merged, df_user, on='Country', how='left')
df_merged.head()
```

```python
df_merged = pd.merge(df_merged, df_market, on='Country', how='left')
df_merged.head()
```

```python
# 定义一个函数，用于每个 Subregion 填充 NaN 值
def fill_na_with_subregion_mean(df, group_col, fill_col):
    # 计算每个 Subregion 的均值
    subregion_means = df.groupby(group_col)[fill_col].transform('mean')
    
    # 使用 Subregion 的均值填充 NaN
    df[fill_col] = df[fill_col].fillna(subregion_means)

# 对所有需要填充的列使用该函数
columns_to_fill = ['u2020', 'u2021', 'u2022', 'u2023', 'm2020', 'm2021', 'm2022', 'm2023',]  # 替换为需要填充的列名

for column in columns_to_fill:
    fill_na_with_subregion_mean(df_merged, 'Subregion', column)
```

```python
na_rows = df_merged[df_merged.isna().any(axis=1)]
na_rows
```
